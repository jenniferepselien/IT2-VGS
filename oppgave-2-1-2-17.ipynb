```python
def trekanttall(n):
    tn = n * (n + 1) // 2
    return tn

total_sum = 0

for i in range(1, 11):
    trekant_tall = trekanttall(i)
    total_sum += trekant_tall

print("Total sum av de ti første trekanttallene er:", total_sum)
```

```python
# 2.14 

def byttPlass(a, i, j):
    temp = a[i]
    a[i] = a[j]
    a[j] = temp

a = [8, 5, 2, 6, 12]
n = len(a)

i = 0
while i < n - 1:
    byttet = False
    for j in range(0, n - i - 1):
        if a[j] > a[j+1]:
            byttPlass(a, j, j+1)
            byttet = True
    if not byttet:
        break
    i += 1

print("Sortert liste:", a)
```

```python
# 2.15
print("-- Huskelista --")

# Les inn fil
with open("huskeliste.txt", "r") as file:
    huskelisten = [line.strip() for line in file.readlines()]

brukervalg = ""
while brukervalg.lower() != "avslutt":
    for gjøremål in huskelisten:
        print(f"- {gjøremål}")
    print("Skriv nytt gjøremål for å legge til i huskelista, avslutt avslutter programmet")
    brukervalg = input("> ")
    if brukervalg.lower() != "avslutt":
        huskelisten.append(brukervalg)

print("Avslutter")

# Skriv til fil
with open("huskeliste.txt", "w") as file:
    for gjøremål in huskelisten:
        file.write(gjøremål + "\n")
```
